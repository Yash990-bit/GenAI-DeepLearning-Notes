# Gradient Descent

optimization algorithm used in Machine Learning to find the minimum value of a function (usually the loss or cost function).


# Goal: Minimize the cost function

It is widely used in models like Linear Regression, Logistic Regression, and Neural Networks.

# It is technique of local minima
(it is backbone of deep learning)



# Intuition of Gradient Descent

Example

Imagine you are standing on a mountain in the dark 🌙 and you want to reach the lowest point (valley).

You cannot see the whole mountain, so what do you do?

You check the slope around your feet.

Move a small step downward.

Check the slope again.

Move downward again.

You repeat this until you reach the lowest point.

# Why small steps?

If you take very big steps, you might jump past the valley.

If you take very tiny steps, it will take forever.

That is why we use a learning rate (α).

Gradient Descent = Repeatedly adjust parameters in the direction that reduces error the fastest.

# Error high → adjust weights
# Error lower → adjust again
# Error minimum → stop

# When Do We Stop?

1️⃣ Parameter change becomes very small

b_new − b_old ≈ 0.0001

Meaning the parameter is not changing anymore.

2️⃣ Fixed number of iterations

Iteration = 1000

epochs

Meaning we train for a fixed number of steps.

# Formula

bnew​=bold​-slope

b_old → current value of parameter

b_new → updated value

slope → gradient (derivative)

*  If b is negative it increases

*  If b is positive it decreases

new value = old value − η × slope

η (eta) → learning rate


# Super Simple Intuition (Best Way to Teach)

Imagine a ball rolling down a bowl.

Steps:

1️⃣ Check slope

2️⃣ Move a little down

3️⃣ Check slope again

4️⃣ Move again

Eventually the ball reaches the bottom.

That is exactly what Gradient Descent does.

# Code

```python
from sklearn.datasets import make_regression
import numpy as np
```

```python
X,y = make_regression(n_samples=4 , n_features=1 , n_informative=1 , n_targets=1 , noise = 80 , random_state=13)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
# Lets apply OLS (Ordinary least square)
from sklearn.linear_model import LinearRegression
```

```python
reg = LinearRegression()
reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red')
```

```python
# Lets apply Gradient Descent slope is constant m = 78.35
# and lets assume intercept b = 0

y_pred = ((78.35*X)+0).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred,color = 'green',label = 'GD')
plt.legend()
plt.show()
```

```python
m= 78.35
b = 0

loss_slope = -2 * np.sum(y-m*X.ravel()-b)
loss_slope
```

```python
# Lets take learning rate = 0.1

lr = 0.1

step_size= loss_slope*lr
step_size
```

```python
# Calculate the new intercept

b = b-step_size
b

```

```python
y_pred1 = ((78.35*X)+b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color = 'red',label = 'OLS')
plt.plot(X,y_pred,color = 'green',label = 'GD')
plt.plot(X,y_pred1,color = 'yellow', label = 'GD'.format(b))
plt.legend()
plt.show()
```

# Effect of Loss Function

L = summation (yi - y^i)^2

it is convex function

A loss function measures the error between:

Actual value (true output)

Predicted value (model output)


# It Determines Model Training Direction

If prediction is too high → decrease parameter

If prediction is too low → increase parameter





 # In Special Case (Weight Update)

 In models like Linear Regression, we update weights and bias separately.

 # Weight update:

          w=w−η*∂L/∂w

 # Bias update:

          b=b−η*∂L​/∂b

  

  # Step Downhill Formula means updating parameters by subtracting the gradient so the error decreases step by step.



# Remember the flow of Gradient Descent

1️⃣ Initialize Parameters

Start with random values for weights and bias.

w = 0

b = 0

2️⃣ Make Prediction

Use the model equation.

ypred​=wx+b

3️⃣ Compute Loss (Error)

Calculate how wrong the prediction is using a loss function.

Loss=(y−ypred​)2

# Big loss = model is very wrong.


4️⃣ Compute Gradient (Slope)

Find how the loss changes with respect to parameters.

∂Loss / ∂w

∂Loss / ∂b

5️⃣ Update Parameters (Step Downhill)

Use the step downhill formula from Gradient Descent.

 w=w−η*∂L/∂w

 b=b−η*∂L​/∂b


  # Meaning:
  new value = old value − learning rate × gradient

   This reduces the error.

   # Then repeat the process until loss become very small.

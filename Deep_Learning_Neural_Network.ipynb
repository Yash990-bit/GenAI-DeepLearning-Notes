```python
# Five point to remember

1. It like a brain for computer

2. Layers: Input → Hidden → Output

3. Neurons = tiny decision-makers

4. Learning happens by trial and error

5. It can solve complex problems


'''Input layer: takes in data.

Hidden layer(s): processes data with weights, bias, and activation functions.

Output layer: gives prediction.'''

Each neuron multiplies inputs by weights, adds bias, applies activation function.
```
